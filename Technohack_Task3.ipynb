```python
import matplotlib.pyplot as plt
plt.hist(df['SepalLengthCm'], bins=15, edgecolor='k')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Sepal Length')
plt.show()
```

```python
plt.hist(df['SepalWidthCm'], bins=15, edgecolor='k')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Sepal Width')
plt.show()
```

```python
plt.hist(df['PetalLengthCm'], bins=15, edgecolor='k')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Petal Length')
plt.show()
```

```python
plt.hist(df['PetalWidthCm'], bins=15, edgecolor='k')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Petal Width')
plt.show()
```

```python
import seaborn as sns
sns.countplot(data=df, x='Species', palette='Set2')
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Distribution of Species')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.barplot(data=df, x='Species', y='SepalLengthCm', palette='Set2')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.title('Sepal Length by Species')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.barplot(data=df, x='Species', y='PetalLengthCm', palette='Set2')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.title('Petal Length by Species')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a violin plot for Sepal Width by Species
plt.figure(figsize=(10, 6))
sns.violinplot(x='Species', y='SepalWidthCm', data=df, palette='Set2')
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.title('Violin Plot of Sepal Width by Species')
plt.show()
```
